```python
slots = range(1,7)
sets = range(1,9)
```

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
from pprint import pprint
```

```python
mods_df = pd.DataFrame()
for each_slot in slots:
    for each_set in sets:
        get_url = f'https://swgoh.gg/p/645186978/mods/?slot_{each_slot}=1&set_{each_set}=1'
        html = requests.get(get_url)
        soup = bs(html.text, 'html.parser')
        mods = soup.find_all('div', class_='collection-mod')
        for each_mod in mods:
            mod_stats = [each_set,each_slot]
            on_character = each_mod.find('img',class_='char-portrait-img')['alt']
            stars = len(each_mod.find_all('span',class_='statmod-pip'))
            mod_stats.append(stars)
            level = each_mod.find('span',class_='statmod-level').text
            mod_stats.append(level)
#             print(level)
            primary = each_mod.find('div',class_='statmod-stats-1').text[2:-1]
            mod_stats.append(primary)
#             print('+'+primary)
            secondaries = each_mod.find('div',class_='statmod-stats-2')
            each_secondary = secondaries.find_all('div',class_='statmod-stat')
        #     print(each_secondary[0])
            for each_sec in each_secondary:
                secondary_value = each_sec.find('span',class_='statmod-stat-value').text[1:]
                secondary_type = each_sec.find('span',class_='statmod-stat-label').text
                secondary_rolls = each_sec.find('span',class_='statmod-stat-roll').text[2]
#                 print(f'this mod has +{secondary_value} {secondary_type} and rolled {secondary_rolls} times')
                mod_stats.append(secondary_value)
                mod_stats.append(secondary_type)
                mod_stats.append(secondary_rolls)
            mod_stats.append(on_character)
            mods_df = mods_df.append(pd.Series(mod_stats),ignore_index=True)

mods_df.columns = ['set','slot','stars','level','primary','sec1-val','sec1-type','sec1-roll',
                                       'sec2-val','sec2-type','sec2-roll',
                                       'sec3-val','sec3-type','sec3-roll',
                                       'sec4-val','sec4-type','sec4-roll','character']
```

```python
def sum_mod_stats(mod_list):
    stat_speed = 0
    stat_crit = 0
    stat_potency = 0
    stat_tenacity = 0
    stat_prot = 0
    stat_prot_pct = 0
    stat_off = 0
    stat_off_pct = 0
    stat_health = 0
    stat_health_pct = 0
    stat_def = 0
    stat_def_pct = 0
    stat_cd = 0
    for mod_slot in range(1,7):
        #primaries:
        primary_type = test_df3.iloc[mod_list[mod_slot]]['primary-type']
        primary_value = test_df3.iloc[mod_list[mod_slot]]['primary-val']
        if primary_type == 'Speed':
            stat_speed += primary_value
        elif primary_type == 'Offense':
            stat_off_pct += primary_value
        elif primary_type == 'Critical Damage':
            stat_cd += primary_value
        elif primary_type == 'Critical Chance':
            stat_crit += primary_value
        elif primary_type == 'Protection':
            stat_prot_pct += primary_value
        elif primary_type == 'Health':
            stat_health_pct += primary_value
        elif primary_type == 'Defense':
            stat_def_pct += primary_value
        elif primary_type == 'Tenacity':
            stat_tenacity += primary_value
        elif primary_type == 'Potency':
            stat_potency += primary_value
        
        #secondaries
        (a,b,c,d,e,f,g,h,i,j,k,l) = test_df3.iloc[mod_list[mod_slot]][['speed-val', 'critical_chance-val', 'potency-val', 'tenacity-val', 'protection-val', 'protection%-val','offense-val','offense%-val','health-val', 'health%-val','defense-val','defense%-val']].fillna(0)
        stat_speed += a
        stat_crit += b
        stat_potency += c
        stat_tenacity += d
        stat_prot += e
        stat_prot_pct += f
        stat_off += g
        stat_off_pct += h
        stat_health += i
        stat_health_pct += j
        stat_def += k
        stat_def_pct += l
    mod_stats = {'speed':stat_speed,'cc':stat_crit,'potency':stat_potency,'tenacity':stat_tenacity,'protection':stat_prot,'protection%':stat_prot_pct,
                'offense':stat_off,'offense%':stat_off_pct,'health':stat_health,'health%':stat_health_pct,'defense':stat_def,'defense%':stat_def_pct,'cd':stat_cd}
    return mod_stats
```
